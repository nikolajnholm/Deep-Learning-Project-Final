```python
import cv2
from PIL import Image
from os import listdir
from os.path import isfile, join
import re
import shutil
import numpy as np
```

## Import highframes

```python
def import_lowframe(path_video, path_save_frames):
    vidcap = cv2.VideoCapture(path_video)
    success, image = vidcap.read()
    count = 0
    transposted_image = np.flip(np.moveaxis(image, 0,1),0)
    
    while success:
        success,image = vidcap.read()
#        if count % 5 == 0:
        cv2.imwrite(path_save_frames + "frame%d.jpg" % count, transposted_image)     # save frame as JPEG file 
        transposted_image = np.flip(np.moveaxis(image, 0,1),0)
        print('Read a new frame: ', success)
        count += 1
```

```python
path_video = "BikeVideo/Video/FILE0004.MP4"
path_save_frames = "BikeVideo/BikeHighframe/"

import_lowframe(path_video, path_save_frames)
```

## Make lowframes

```python
jpgpath = 'BikeVideo/BikeHighframe/'

jpgfiles = [f for f in listdir(jpgpath) if isfile(join(jpgpath, f))]
jpgfiles.sort()
jpgfiles = jpgfiles[1:]
```

```python
p = '[\d]+[.,\d]+|[\d]*[.][\d]+|[\d]+'
img_lowframe = []

for img in jpgfiles:
    num_frame = img.split(".")[0]
    if int(re.findall(p, num_frame)[0]) % 5 == 0:
        img_lowframe.append(jpgpath + img)
```

```python
for f in img_lowframe:
    shutil.copy(f, 'BikeVideo/Bikelowframe')
```
