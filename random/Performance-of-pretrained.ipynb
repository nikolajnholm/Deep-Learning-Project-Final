```python
pkl_data = reader.load_pickle("/Users/oskarlassen/Desktop/DeepLearning/Project/Deep-Learning-Project/Annotations/Frames_forannotation.pkl")
```

```python
img_dic = {}

for i in range(pkl_data.shape[0]):
    ####### MAKING PICKLE IMAGE READY TO GO INTO YOLO
    img_pkl = pkl_data[i]
    # scale pixel values to [0, 1]
    img = img_pkl.astype('float32')
    img /= 255.0
    # add a dimension so that we have one sample
    img = expand_dims(img, 0)
    
    ####### RETURN ########

    labels = YOLO_pkl_predict(img, model)
    img_dic[i] = labels
    if i % 200 == 0:
        print(i)
```
